# Import requests to make http requests

```python
import requests
```

# Get request to url of the website

```python
response = requests.get(f"https://books.toscrape.com")
print(response.text)
```

# Check if status code is 200

```python
response =requests.get(f"https://books.toscrape.com")
if response.status_code == 200:
    print(response.text)
```

# Request was unsuccessful

```python
response =requests.get(f"https://books.example.com")
if response.status_code == 200:
    print(response.text)
```

# Using try and except to make the error look good

```python
try:
    response =requests.get(f"https://books.example.com")
    if response.status_code == 200:
        print(response.text)
except requests.exceptions.RequestException as e:
    print("Failed to retrieve the webpage:")
```

# Checking common status codes

```python
response =requests.get(f"https://books.toscrape.com/catalogue/page-51.html")
print(response.status_code)
if response.status_code == 200:
    print(response.text)
elif response.status_code == 404:
    print("Page not found")
elif response.status_code == 403:
    print("Access forbidden")
elif response.status_code == 500:
    print("Internal server error")
```

# Using loop to get all the pages

```python
for i in range(1, 51):
    response =requests.get(f"https://books.toscrape.com/catalogue/page-{i}.html")
    if response.status_code == 200:
        print(f"Page {i} Downloaded successfully.")
        print("Page Content:")
        print(' '.join(response.text.split())[:100])
        
```

# Using while loop as number of pages is unknown

```python
i = 1
while True :
    response =requests.get(f"https://books.toscrape.com/catalogue/page-{i}.html")
    if response.status_code != 200:
        break # Exit the loop if the page is not found
    else:
        print(f"Page {i} Downloaded successfully.")
        print("Page Content:")
        print(' '.join(response.text.split())[:100])
        i += 1
```

# Storing html in files

```python
import os

os.makedirs("html", exist_ok=True)

i = 1
while True :
    
    response =requests.get(f"https://books.toscrape.com/catalogue/page-{i}.html")
    if response.status_code != 200:
        break
    else:
        print(f"Page {i} Downloaded successfully.")
        with open(f"html/page_{i}.html", "w", encoding="utf-8") as file:
            file.write(response.text)
        i += 1
```
